```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
```

```python
model = tf.keras.models.load_model('mobilenet2/')
```

```python
seq = list()

img = image.load_img('pic.jpeg', target_size=(224, 224))
img = image.img_to_array(img,dtype=np.float16)
img = preprocess_input(img)
seq.append(img)
```

```python
seq = np.array(seq,dtype=np.float32)
```

# Conversion

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
```

```python
tflite_model = converter.convert()
```

```python
interpreter = tf.lite.Interpreter(model_content=tflite_model)
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
expected = model.predict(seq)
```

```python
interpreter.set_tensor(input_details[0]["index"], seq)   # seq is the input
interpreter.invoke()
result = interpreter.get_tensor(output_details[0]["index"])
```
